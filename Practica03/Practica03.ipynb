Campo Descripcion <br>
**#** El ID o numero de identificacion del Pokemo
**Name:** El nombre del Pokémon <br>
**Type 1:** Se refiere a las partes elementales (tipos) asociados tanto al Pokémon como a sus movimientos. <br>
**Type 2:** Los Pokémon pueden tener hasta 2 tipos (Pokémon de tipo dual); los movimientos solo pueden tener uno. <br>
**Total:** Representa la suma de todos los stats, lo que te da una idea de qué tan “pesado” o fuerte es el Pokémon. <br>
**HP** (Puntos de Salud): Es qué tanto aguante tiene un Pokémon antes de que lo dejen fuera de combate (se desmaya). <br>
**Attack:** La fuerza física de los ataques del Pokémon. <br>
**Defense:** Qué tanto aguanta el Pokémon contra los ataques físicos del rival. <br>
**Sp. Atk:** Ataque Especial; el poder de los movimientos especiales del Pokémon. <br>
**Sp. Def:** Defensa Especial; qué tan resistente es el Pokémon ante los ataques especiales. <br>
**Speed** (Velocidad): El Pokémon que tenga más velocidad ataca primero. <br>
**Generations:** Se refiere a la generación de la serie de juegos a la que pertenece. <br>
**Legendary:** Indica si el Pokémon es Legendario o no (son bichos muy raros y poderosos). Es un valor booleano (sí/no). <br>
