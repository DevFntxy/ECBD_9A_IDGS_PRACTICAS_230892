15. Gráfico de distribución de Pokémon por tipo

Se generó un gráfico de barras utilizando la función countplot() de Seaborn para visualizar la cantidad de Pokémon pertenecientes a cada tipo, con un tamaño de figura de 15 × 5 pulgadas para mejorar la distribucion

16. Analizar la distribución de las estadísticas principales de los Pokémon mediante diagramas de caja.

Se generaron boxplots para las variables numéricas de combate (HP, Attack, Defense, Sp. Atk, Sp. Def y Speed). Se excluyeron las columnas de identificación y las variables categóricas que no aportan información a este análisis.

```python
sns.set_style('darkgrid')
plt.figure(figsize=(10,6))
sns.boxplot(data=pokemon.drop(['#','Total','Generation','Legendary'],axis=1),fliersize=3,palette='seismic')
plt.title('Boxplots for stats')
```

17. Gráfico de violín de las estadísticas de los Pokémon

Primero se crea una figura de 10 × 6 pulgadas para mejorar la visualización de los datos.

Posteriormente, se utiliza la función violinplot() de Seaborn para representar la distribución de las estadísticas numéricas de los Pokémon. Para ello, se eliminan las columnas **#** , **Total, Generation** y **Legendary**, ya que no forman parte de las estadísticas individuales que se desean analizar.

La paleta de colores rocket se emplea para diferenciar visualmente cada variable representada

```python
plt.figure(figsize=(10,6))
sns.violinplot(data=pokemon.drop(['#','Total','Generation','Legendary'],axis=1),palette='rocket')
plt.title('Violinplots for stats')
```

18. Agrupación de estadísticas por tipo de Pokémon

Las estadísticas tienen una distribución similar,Visualicemos los Pokémon agrupados por tipo.

```python
pokemon.groupby('Type').sum()
```

19. Obtención de la suma total de HP por tipo de Pokémon

Se agruparon los registros del conjunto de datos utilizando la variable **Type**. Posteriormente, se sumaron las estadísticas numéricas de cada grupo mediante la función **sum()**. Finalmente, se seleccionó la columna **HP** para obtener una serie que muestra la suma total de puntos de salud para cada tipo de Pokémon.

```python
pokemon.groupby('Type').sum().HP
# En esta serie, los tipos están ordenados alfabéticamente.
```
