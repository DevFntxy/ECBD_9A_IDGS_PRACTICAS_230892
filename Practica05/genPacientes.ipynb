```python
import pandas as pd
import numpy as np
import random

# ======================================================
# CONFIGURACIÓN
# ======================================================

NUM_REGISTROS = 5000

random.seed(42)
np.random.seed(42)

# ======================================================
# PRINCIPALES CIUDADES DE PUEBLA
# (Latitud, Longitud, Peso poblacional aproximado)
# ======================================================

ciudades = [
    ("Puebla",              19.0414, -98.2063, 0.35),
    ("Tehuacán",            18.4665, -97.3987, 0.12),
    ("San Pedro Cholula",   19.0644, -98.3035, 0.09),
    ("Atlixco",             18.9081, -98.4364, 0.07),
    ("San Martín Texmelucan",19.2834,-98.4389, 0.06),
    ("Amozoc",              19.0455, -98.0453, 0.05),
    ("Teziutlán",           19.8186, -97.3600, 0.05),
    ("Huauchinango",        20.1744, -98.0538, 0.04),
    ("Zacatlán",            19.9337, -97.9615, 0.04),
    ("Izúcar de Matamoros", 18.6016, -98.4650, 0.04),
    ("Xicotepec",           20.2764, -97.9609, 0.03),
    ("Libres",              19.4657, -97.6870, 0.03),
    ("Tecamachalco",        18.8834, -97.7289, 0.03)
]

pesos = [c[3] for c in ciudades]

# ======================================================
# GENERAR COORDENADAS
# ======================================================

def generar_coordenadas():
    ciudad = random.choices(ciudades, weights=pesos, k=1)[0]

    _, lat_base, lon_base, _ = ciudad

    lat = np.random.normal(lat_base, 0.06)
    lon = np.random.normal(lon_base, 0.06)

    lat = min(max(lat, 18.15), 20.85)
    lon = min(max(lon, -98.75), -96.70)

    return round(lat, 6), round(lon, 6)

# ======================================================
# CÁLCULO DEL RIESGO
# ======================================================

def calcular_riesgo(
        edad,
        imc,
        sistolica,
        glucosa,
        colesterol,
        trigliceridos,
        fumador,
        diabetes,
        hipertension,
        actividad,
        antecedente):

    puntos = 0

    if edad >= 65:
        puntos += 3
    elif edad >= 50:
        puntos += 2
    elif edad >= 40:
        puntos += 1

    if imc >= 30:
        puntos += 2
    elif imc >= 25:
        puntos += 1

    if sistolica >= 140:
        puntos += 2
    elif sistolica >= 130:
        puntos += 1

    if glucosa >= 126:
        puntos += 2

    if colesterol >= 240:
        puntos += 2
    elif colesterol >= 200:
        puntos += 1

    if trigliceridos >= 200:
        puntos += 1

    if fumador == "Sí":
        puntos += 2

    if diabetes == "Sí":
        puntos += 2

    if hipertension == "Sí":
        puntos += 2

    if antecedente == "Sí":
        puntos += 1

    if actividad == "Baja":
        puntos += 1

    if puntos <= 3:
        return "Bajo"
    elif puntos <= 8:
        return "Medio"
    else:
        return "Alto"

# ======================================================
# GENERACIÓN DEL DATASET
# ======================================================

datos = []

for i in range(1, NUM_REGISTROS + 1):

    edad = random.randint(18, 90)

    sexo = random.choice(["Masculino", "Femenino"])

    peso = round(random.uniform(45, 125), 1)

    estatura = round(random.uniform(1.45, 1.95), 2)

    imc = round(peso / (estatura ** 2), 2)

    sistolica = random.randint(90, 190)

    diastolica = random.randint(60, 110)

    frecuencia = random.randint(55, 120)

    glucosa = random.randint(70, 250)

    colesterol = random.randint(120, 330)

    trigliceridos = random.randint(60, 350)

    oxigenacion = random.randint(90, 100)

    fumador = np.random.choice(["Sí", "No"], p=[0.28, 0.72])

    diabetes = np.random.choice(["Sí", "No"], p=[0.18, 0.82])

    hipertension = np.random.choice(["Sí", "No"], p=[0.25, 0.75])

    actividad = np.random.choice(
        ["Alta", "Media", "Baja"],
        p=[0.30, 0.45, 0.25]
    )

    antecedente = np.random.choice(["Sí", "No"], p=[0.35, 0.65])

    latitud, longitud = generar_coordenadas()

    riesgo = calcular_riesgo(
        edad,
        imc,
        sistolica,
        glucosa,
        colesterol,
        trigliceridos,
        fumador,
        diabetes,
        hipertension,
        actividad,
        antecedente
    )

    datos.append([
        i,
        edad,
        sexo,
        peso,
        estatura,
        imc,
        sistolica,
        diastolica,
        frecuencia,
        glucosa,
        colesterol,
        trigliceridos,
        oxigenacion,
        fumador,
        diabetes,
        hipertension,
        actividad,
        antecedente,
        latitud,
        longitud,
        riesgo
    ])

columnas = [
    "id_paciente",
    "edad",
    "sexo",
    "peso_kg",
    "estatura_m",
    "imc",
    "presion_sistolica",
    "presion_diastolica",
    "frecuencia_cardiaca",
    "glucosa",
    "colesterol",
    "trigliceridos",
    "oxigenacion",
    "fumador",
    "diabetes",
    "hipertension",
    "actividad_fisica",
    "antecedente_familiar",
    "latitud",
    "longitud",
    "riesgo"
]

df = pd.DataFrame(datos, columns=columnas)

df.to_csv("pacientes_puebla.csv", index=False, encoding="utf-8-sig")

print(df.head())
print("\nDimensiones:", df.shape)
print("\nDistribución del riesgo:")
print(df["riesgo"].value_counts())
```
