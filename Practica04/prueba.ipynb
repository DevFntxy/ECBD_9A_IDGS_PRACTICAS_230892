# A Visual and Insightful Journey - DonorsChoose

## Análisis Exploratorio y Visualización de Datos

En este notebook realizaremos un análisis completo del dataset de DonorsChoose, explorando proyectos, donantes, escuelas y recursos para obtener insights valiosos sobre la plataforma.

**Contenidos:**
1. Carga y exploración de datos
2. Análisis descriptivo
3. Visualizaciones por categoría
4. Análisis geográfico
5. Correlaciones y patrones

## 1. Importar librerías necesarias

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import warnings
warnings.filterwarnings('ignore')

# Configuración de estilos
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (14, 6)
plt.rcParams['font.size'] = 10

print("Librerías cargadas exitosamente")
```

## 2. Cargar datos

```python
import os

# Ruta base
base_path = r'c:\Users\derek\Desktop\ECBD_9A_IDGS_PRACTICAS_230892\Practica04'

# Cargar datasets
projects = pd.read_csv(os.path.join(base_path, 'Projects.csv'))
donations = pd.read_csv(os.path.join(base_path, 'Donations.csv'))
donors = pd.read_csv(os.path.join(base_path, 'Donors.csv'))
teachers = pd.read_csv(os.path.join(base_path, 'Teachers.csv'))
schools = pd.read_csv(os.path.join(base_path, 'Schools.csv'))
resources = pd.read_csv(os.path.join(base_path, 'Resources.csv'))

print("Datasets cargados:\n")
print(f"Projects: {projects.shape}")
print(f"Donations: {donations.shape}")
print(f"Donors: {donors.shape}")
print(f"Teachers: {teachers.shape}")
print(f"Schools: {schools.shape}")
print(f"Resources: {resources.shape}")
```

## 3. Exploración Inicial de Proyectos

```python
print("PRIMEROS REGISTROS DE PROYECTOS:")
print(projects.head())
print("\nINFORMACIÓN DE COLUMNAS:")
print(projects.info())
print("\nESTADÍSTICAS DESCRIPTIVAS:")
print(projects.describe())
```

## 4. Análisis de Categorías de Proyectos

```python
# Análisis de categorías si existen
if 'Project_Category_Code' in projects.columns:
    category_counts = projects['Project_Category_Code'].value_counts()
    
    fig, ax = plt.subplots(figsize=(12, 6))
    category_counts.plot(kind='barh', ax=ax, color='steelblue')
    ax.set_title('Cantidad de Proyectos por Categoría', fontsize=14, fontweight='bold')
    ax.set_xlabel('Cantidad de Proyectos')
    ax.set_ylabel('Categoría')
    plt.tight_layout()
    plt.show()
    
    print(f"\nTop 5 Categorías:\n{category_counts.head()}")
else:
    print("Columna 'Project_Category_Code' no encontrada")
    print(f"Columnas disponibles: {projects.columns.tolist()}")
```

## 5. Análisis de Financiamiento

```python
# Análisis de montos
print("ESTADÍSTICAS DE FINANCIAMIENTO:")
print(f"\nProyectos por estado de financiamiento:")
if 'Project_Status' in projects.columns:
    print(projects['Project_Status'].value_counts())
    
    fig, ax = plt.subplots(figsize=(10, 6))
    projects['Project_Status'].value_counts().plot(kind='pie', ax=ax, autopct='%1.1f%%', colors=['#2ecc71', '#e74c3c', '#f39c12'])
    ax.set_title('Distribución de Estados de Proyectos', fontsize=14, fontweight='bold')
    ax.set_ylabel('')
    plt.tight_layout()
    plt.show()

# Visualizar monto de financiamiento si existe
numeric_cols = projects.select_dtypes(include=[np.number]).columns.tolist()
print(f"\nColumnas numéricas disponibles: {numeric_cols}")
```

## 6. Análisis de Donaciones

```python
print("INFORMACIÓN DE DONACIONES:")
print(f"Total de donaciones: {len(donations)}")
print(f"\nPrimeros registros:")
print(donations.head())
print(f"\nEstadísticas:")
print(donations.describe())

# Análisis de monto de donación
if 'Donation_Amount' in donations.columns or 'Donation' in donations.columns:
    amount_col = 'Donation_Amount' if 'Donation_Amount' in donations.columns else 'Donation'
    
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    
    # Histograma
    axes[0].hist(donations[amount_col].dropna(), bins=50, color='steelblue', edgecolor='black')
    axes[0].set_title('Distribución de Montos de Donación', fontsize=12, fontweight='bold')
    axes[0].set_xlabel('Monto de Donación ($)')
    axes[0].set_ylabel('Frecuencia')
    
    # Box plot
    axes[1].boxplot(donations[amount_col].dropna())
    axes[1].set_title('Box Plot de Montos de Donación', fontsize=12, fontweight='bold')
    axes[1].set_ylabel('Monto de Donación ($)')
    
    plt.tight_layout()
    plt.show()
    
    print(f"\nEstadísticas del monto de donación:")
    print(f"Monto promedio: ${donations[amount_col].mean():.2f}")
    print(f"Monto mediano: ${donations[amount_col].median():.2f}")
    print(f"Monto máximo: ${donations[amount_col].max():.2f}")
    print(f"Monto mínimo: ${donations[amount_col].min():.2f}")
```

## 7. Análisis de Donantes

```python
print("INFORMACIÓN DE DONANTES:")
print(f"Total de donantes: {len(donors)}")
print(f"\nPrimeros registros:")
print(donors.head())
print(f"\nColumnas: {donors.columns.tolist()}")

# Análisis de donantes por ubicación si existe
if 'Donor_State' in donors.columns:
    state_counts = donors['Donor_State'].value_counts().head(15)
    
    fig, ax = plt.subplots(figsize=(12, 7))
    state_counts.plot(kind='barh', ax=ax, color='coral')
    ax.set_title('Top 15 Estados con más Donantes', fontsize=14, fontweight='bold')
    ax.set_xlabel('Cantidad de Donantes')
    plt.tight_layout()
    plt.show()
```

## 8. Análisis de Escuelas

```python
print("INFORMACIÓN DE ESCUELAS:")
print(f"Total de escuelas: {len(schools)}")
print(f"\nPrimeros registros:")
print(schools.head())
print(f"\nColumnas: {schools.columns.tolist()}")

# Análisis por estado si existe
if 'School_State' in schools.columns:
    school_state = schools['School_State'].value_counts().head(15)
    
    fig, ax = plt.subplots(figsize=(12, 7))
    school_state.plot(kind='barh', ax=ax, color='skyblue')
    ax.set_title('Top 15 Estados con más Escuelas', fontsize=14, fontweight='bold')
    ax.set_xlabel('Cantidad de Escuelas')
    plt.tight_layout()
    plt.show()

# Análisis por tipo de escuela si existe
if 'School_Type' in schools.columns:
    school_type = schools['School_Type'].value_counts()
    
    fig, ax = plt.subplots(figsize=(10, 6))
    school_type.plot(kind='pie', ax=ax, autopct='%1.1f%%', startangle=90)
    ax.set_title('Distribución por Tipo de Escuela', fontsize=14, fontweight='bold')
    ax.set_ylabel('')
    plt.tight_layout()
    plt.show()
```

## 9. Análisis de Recursos

```python
print("INFORMACIÓN DE RECURSOS:")
print(f"Total de recursos: {len(resources)}")
print(f"\nPrimeros registros:")
print(resources.head())
print(f"\nColumnas: {resources.columns.tolist()}")

# Análisis de cantidad de recursos por proyecto si existe
if 'Quantity' in resources.columns:
    fig, ax = plt.subplots(figsize=(12, 6))
    resources['Quantity'].value_counts().head(15).plot(kind='bar', ax=ax, color='mediumpurple')
    ax.set_title('Top 15 Cantidades de Recursos más Frecuentes', fontsize=14, fontweight='bold')
    ax.set_xlabel('Cantidad')
    ax.set_ylabel('Frecuencia')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

## 10. Análisis de Maestros

```python
print("INFORMACIÓN DE MAESTROS:")
print(f"Total de maestros: {len(teachers)}")
print(f"\nPrimeros registros:")
print(teachers.head())
print(f"\nColumnas: {teachers.columns.tolist()}")

# Estadísticas descriptivas
print(f"\nEstadísticas:")
print(teachers.describe())
```

## 11. Análisis de Correlaciones (Proyectos)

```python
# Seleccionar solo columnas numéricas para correlación
numeric_cols = projects.select_dtypes(include=[np.number]).columns.tolist()

if len(numeric_cols) > 1:
    correlation_matrix = projects[numeric_cols].corr()
    
    fig, ax = plt.subplots(figsize=(10, 8))
    sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', center=0, ax=ax)
    ax.set_title('Matriz de Correlaciones - Proyectos', fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.show()
    
    print("Matriz de Correlaciones:")
    print(correlation_matrix)
else:
    print(f"Columnas numéricas encontradas: {numeric_cols}")
    print("No hay suficientes columnas numéricas para análisis de correlación")
```

## 12. Análisis Combinado: Proyectos y Donaciones

```python
# Merge de datos para análisis combinado
print("Analizando relación entre proyectos y donaciones...")

# Verificar columnas comunes
print(f"\nColumnas en Projects: {projects.columns.tolist()}")
print(f"Columnas en Donations: {donations.columns.tolist()}")

# Intentar merge si hay columna común
common_cols = set(projects.columns) & set(donations.columns)
print(f"\nColumnas comunes: {common_cols}")

if len(common_cols) > 0:
    merge_col = list(common_cols)[0]
    print(f"\nRealizando merge por: {merge_col}")
    
    # Merge
    merged = projects.merge(donations, on=merge_col, how='left')
    print(f"Dataset combinado: {merged.shape}")
    print(f"\nPrimeros registros del merge:")
    print(merged.head())
else:
    print("\nNo hay columnas comunes para hacer merge automático")
    print("Análisis separado completado")
```

## 13. Resumen de Insights Clave

```python
print("=" * 70)
print("RESUMEN DE INSIGHTS CLAVE - DONORSCHOOSE")
print("=" * 70)

print("\n📊 ESTADÍSTICAS GENERALES:")
print(f"  • Total de Proyectos: {len(projects):,}")
print(f"  • Total de Donaciones: {len(donations):,}")
print(f"  • Total de Donantes: {len(donors):,}")
print(f"  • Total de Escuelas Participantes: {len(schools):,}")
print(f"  • Total de Maestros: {len(teachers):,}")
print(f"  • Total de Recursos Solicitados: {len(resources):,}")

print("\n💰 ANÁLISIS FINANCIERO:")
if 'Donation_Amount' in donations.columns or 'Donation' in donations.columns:
    amount_col = 'Donation_Amount' if 'Donation_Amount' in donations.columns else 'Donation'
    print(f"  • Monto Total Donado: ${donations[amount_col].sum():,.2f}")
    print(f"  • Donación Promedio: ${donations[amount_col].mean():,.2f}")
    print(f"  • Donación Mediana: ${donations[amount_col].median():,.2f}")
    print(f"  • Donación Máxima: ${donations[amount_col].max():,.2f}")

print("\n🎯 ANÁLISIS DE PROYECTOS:")
if 'Project_Status' in projects.columns:
    status_dist = projects['Project_Status'].value_counts()
    for status, count in status_dist.items():
        pct = (count / len(projects)) * 100
        print(f"  • {status}: {count:,} ({pct:.1f}%)")

print("\n🌍 DATOS GEOGRÁFICOS:")
if 'School_State' in schools.columns:
    print(f"  • Estados con Escuelas: {schools['School_State'].nunique()}")
    top_state = schools['School_State'].value_counts().index[0]
    top_count = schools['School_State'].value_counts().values[0]
    print(f"  • Estado con más escuelas: {top_state} ({top_count} escuelas)")

if 'Donor_State' in donors.columns:
    print(f"  • Estados con Donantes: {donors['Donor_State'].nunique()}")

print("\n" + "=" * 70)
print("✅ Análisis completado exitosamente")
print("=" * 70)
```
