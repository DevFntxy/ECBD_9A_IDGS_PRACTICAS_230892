IMPORTAR LIBRERIAS

```python
# ============================================
# LIBRERÍAS
# ============================================

import os
import glob
import warnings
import numpy as np
import pandas as pd

warnings.filterwarnings("ignore")

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 20)
```

DEFINIR RUTAS

```python
# ============================================
# RUTAS
# ============================================

RUTA_DATOS = "../Data/raws/*.csv"

SALIDA = "../Data/processed"

os.makedirs(SALIDA, exist_ok=True)
```

BUSCAR DATASETS

```python
# ============================================
# BUSCAR DATASETS
# ============================================

archivos = sorted(glob.glob(RUTA_DATOS))

print("="*60)
print("DATASETS ENCONTRADOS")
print("="*60)

for archivo in archivos:
    print(os.path.basename(archivo))

print("\nTotal:", len(archivos))
```

CARGAR DATASETS

```python
# ============================================
# CARGAR DATASETS
# ============================================

datasets = {}

for archivo in archivos:

    nombre = os.path.basename(archivo).replace(".csv","")

    datasets[nombre] = pd.read_csv(archivo)

print("Datasets cargados correctamente.")
```

INFORMACION DE DATASET

```python
# ============================================
# INFORMACIÓN
# ============================================

resumen = []

for nombre,df in datasets.items():

    resumen.append({

        "Dataset":nombre,

        "Registros":df.shape[0],

        "Columnas":df.shape[1]

    })

resumen = pd.DataFrame(resumen)

display(resumen)
```

REVISAR ESTRUCTURA

```python
# ============================================
# COLUMNAS
# ============================================

for nombre,df in datasets.items():

    print("="*60)

    print(nombre)

    print("="*60)

    print(df.columns.tolist())
```

ESQUEMA MAESTRO

```python
COLUMNAS = [

"ID_Paciente",

"CURP",

"Nombre",

"Apellido_Paterno",

"Apellido_Materno",

"Fecha_Nacimiento",

"Edad",

"Sexo",

"Estatura_m",

"Peso_kg",

"IMC",

"Presion_Sistolica",

"Presion_Diastolica",

"Glucosa",

"Colesterol",

"Frecuencia_Cardiaca",

"Diabetes",

"Hipertension",

"Fumador",

"Actividad_Fisica",

"Antecedente_Familiar",

"Municipio",

"Region",

"Latitud",

"Longitud"

]
```

DICCIONARIO EQUIVALENTE

```python
RENOMBRAR = {

    "id_paciente":"ID_Paciente",

    "Genero":"Sexo",

    "genero":"Sexo",

    "sexo":"Sexo",

    "edad":"Edad",

    "altura_cm":"Estatura_cm",

    "estatura_cm":"Estatura_cm",

    "estatura_m":"Estatura_m",

    "Peso":"Peso_kg",

    "peso":"Peso_kg",

    "peso_kg":"Peso_kg",

    "Indice_Masa_Corporal":"IMC",

    "imc":"IMC",

    "presion_sistolica":"Presion_Sistolica",

    "Presion_Sistolica_mmHg":"Presion_Sistolica",

    "Presion_Arterial_Sistolica":"Presion_Sistolica",

    "presion_diastolica":"Presion_Diastolica",

    "Presion_Diastolica_mmHg":"Presion_Diastolica",

    "Glucosa_mgdl":"Glucosa",

    "Glucosa_mg_dl":"Glucosa",

    "glucosa":"Glucosa",

    "Colesterol_Total":"Colesterol",

    "Colesterol_mg_dl":"Colesterol",

    "colesterol":"Colesterol",

    "Frecuencia_Cardiaca_lpm":"Frecuencia_Cardiaca",

    "frecuencia_cardiaca":"Frecuencia_Cardiaca",

    "Historial_Familiar":"Antecedente_Familiar",

    "antecedente_familiar":"Antecedente_Familiar",

    "latitud":"Latitud",

    "longitud":"Longitud",

    "municipio":"Municipio"

}
```

FUNCION PARA PREPARAR EL DATASET

```python
def preparar_dataset(df):

    # Renombrar columnas
    df = df.rename(columns=RENOMBRAR)

    # Convertir estatura de centímetros a metros
    if "Estatura_cm" in df.columns:

        df["Estatura_m"] = df["Estatura_cm"] / 100

        df.drop(columns="Estatura_cm", inplace=True)

    # Agregar columnas faltantes
    for columna in COLUMNAS:

        if columna not in df.columns:

            df[columna] = np.nan

    # Mantener únicamente el esquema maestro
    df = df[COLUMNAS]

    return df
```

LIMPIAR TODOS LOS DARASETS

```python
datasets_limpios = []

for nombre,df in datasets.items():

    print("Procesando:",nombre)

    limpio = preparar_dataset(df)

    datasets_limpios.append(limpio)

print("\nTodos los datasets fueron normalizados.")
```

UNIFICAR

```python
if datasets_limpios:
    dataset = pd.concat(datasets_limpios, ignore_index=True)
else:
    dataset = pd.DataFrame(columns=COLUMNAS)

print(dataset.shape)

print(dataset.shape)
```

ELIMINAR DATOS DUPLICADOS

```python
print("Antes:",dataset.shape)

dataset = dataset.drop_duplicates()

print("Después:",dataset.shape)
```

REVISAR DATOS NULOS

```python
nulos = dataset.isnull().sum()

nulos = nulos[nulos>0]

display(

    nulos.sort_values(ascending=False)

)
```

INFORMACION FINAL

```python
print("="*60)

print("DATASET UNIFICADO")

print("="*60)

print("Registros :",dataset.shape[0])

print("Columnas :",dataset.shape[1])

display(dataset.head())
```

GUARDAR DATOS DEL DATASET

```python
dataset.to_csv(

    "data/processed/dataset_unificado.csv",

    index=False,

    encoding="utf-8-sig"

)

print("Dataset guardado correctamente.")
```
